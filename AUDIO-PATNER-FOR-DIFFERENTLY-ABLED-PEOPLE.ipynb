```python
import PyPDF2
from tkinter import *
from pydub import AudioSegment
import os
from googletrans import Translator
from gtts import gTTS

def pdf_to_aud_trans():
    audioclips = []
    translator = Translator()
    with open(path_field.get(), 'rb') as path:
        pdfReader = PyPDF2.PdfReader(path)
        page, page_end = int(page_field.get()), int(page_field2.get())
        audio = direc_field.get()
        lng = clicked.get()
        
        for i in range(page, page_end + 1):
            audiox = f"{audio}{i}"
            text = pdfReader.pages[i-1].extract_text()
            translated_text = translator.translate(text, dest=lng).text
            gTTS(translated_text, lang=lng).save(f"{audiox}.mp3")
            audioclips.append(f"{os.getcwd()}\\{audiox}.mp3")
    
    final_audio = AudioSegment.from_file(audioclips[0], format="mp3")
    for clip in audioclips[1:]:
        final_audio += AudioSegment.from_file(clip, format="mp3")
    final_audio.export(f"{audio}.mp3", format="mp3")
    
    for clip in audioclips:
        os.remove(clip)

def clear_all():
    for field in (path_field, page_field, direc_field, page_field2):
        field.delete(0, END)

if __name__ == "__main__":
    root = Tk()
    root.iconbitmap(r"C:\Users\Aryar\OneDrive\Desktop")
    root.configure(background='#abcacf')
    root.geometry("400x310")
    root.title("PDF TO AUDIO CONVERTER")
    
    labels = [
        ("Enter PDF directory : ", 1),
        ("Enter page to convert : ", 2),
        ("Enter till which page to convert : ", 3),
        ("Enter name to save : ", 4),
        ("Choose language : ", 5)
    ]
    
    for text, row in labels:
        Label(root, text=text, fg='black', bg='#499959').grid(row=row, column=0, padx=10, pady=10)
    
    options = ["en", "hi", "ta", "te", "ch", "ja"]
    clicked = StringVar(value="en")
    
    fields = [Entry(root) for _ in range(4)]
    for i, field in enumerate(fields):
        field.grid(row=i+1, column=1, padx=10, pady=10)
    
    path_field, page_field, page_field2, direc_field = fields
    
    OptionMenu(root, clicked, *options).grid(row=5, column=1, padx=10, pady=10)
    
    Button(root, text="Create Audio", bg="#2c3bb0", fg="white", command=pdf_to_aud_trans).grid(row=8, column=1, pady=10)
    Button(root, text="Clear", bg="#2c3bb0", fg="white", command=clear_all).grid(row=10, column=1, pady=10)
    
    root.mainloop()
```
